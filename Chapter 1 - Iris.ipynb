```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
iris_df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(15, 15), marker='o',
                          hist_kwds={'bins': 20}, s=60, alpha=0.8, cmap=mglearn.cm3)
```

```python
iris_df['species'] = y_train
iris_df['species'] = iris_df['species'].apply(lambda x : iris_dataset['target_names'][x])
```

```python
sns.pairplot(iris_df, hue="species")
```
